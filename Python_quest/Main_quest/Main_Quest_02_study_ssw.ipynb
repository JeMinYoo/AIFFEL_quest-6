```python
# 문자열 정규화

# 1단계: 토큰화
import re

def tokenize(name): # 토큰화 함수 정의
    name_split = name.split(" ")
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
    
    return np.array(tokens)

all_tokens = list(legendary["Name"].apply(tokenize).values) # 판다스 메서드 apply를 통해서 tokenize 함수 적용, value 메서드로 NumPy 배열로 반환후 all_tokens에 리스트로 저장

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter

most_common = Counter(token_set).most_common(10)    # 파이썬 콜렉션 패키지를 활용해서 가장 많이 등장하는 단어 10개를 순서대로 개수를 표기
most_common
```

```python
# 2단계: 전설적인 포켓몬의 이름에서 가장 많이 등장하는 이름 10개 각각에 해당하는 단어를 가지고 있는지를 True, False 값으로 변환하여 추가 칼럼을 생성

for token, _ in most_common:    # most_common의 튜플 리스트가 가진 빈도 값을 무시하기 위해서 _ 언더바를 넣음
    # pokemon[token] = ... 형식으로 사용하면 뒤에서 warning이 발생합니다
    pokemon[f"{token}"] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

```python
# 베이스라인 모델 만들기

print(original_data.shape)
print(original_data.head())
print(original_data.columns)
```

```python
# 문자열 데이터는 제거 & 타겟 테이터(Legendary) 제거(지금은 전설의 포켓몬 찾는 과정중)
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation'] # 모델 설계자가 제거
target = 'Legendary' # 타겟 데이터 칼럼명은 별도 변수명에 저장
```

```python
# Q. 'original_data'에서 'features' 컬럼에 해당하는 데이터를 변수 'X'에 저장합니다.
X = original_data[features]
print(X.shape)
X.head()
```

```python
# Q. 'target' 컬럼의 데이터를 변수 'y'에 저장합니다.
y = original_data[target]
print(y.shape)
y.tail()
```

```python
# 훈련, 검증 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 모델 학습: 의사결정 트리
from sklearn.tree import DecisionTreeClassifier
print('슝=3')
```

```python
model = DecisionTreeClassifier(random_state=25)
model
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# 상식 수준의 기준점
from sklearn.metrics import confusion_matrix
confu_matrix_ratio = confusion_matrix(y_test, y_pred)
print(confu_matrix_ratio)
print(f'컨퓨젼 매트릭스 정확도: {(confu_matrix_ratio[0][0]+confu_matrix_ratio[1][1])/confu_matrix_ratio.sum()}')
print(f'오리지널 데이터에서 일반 포켓몬 맞출 수 있는 확률: {(1-len(legendary)/len(original_data))}')
```

```python
# 성능이 좋으려면, 오리지널 데이터나 컨퓨젼 매트릭스 보다 더 좋은 값이 나와야 한다.

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred)) # recall 값이 0.62로 상당히 낮음. 타겟을 맞추지 못했단 의미임. 즉, 전설의 포켓몬을 일반 포켓몬으로 분류했다는 의미임
```
